```python
# importamos las librerias
import matplotlib.pyplot as plt

## GRAFICA DE BARRAS

#declaramos las variables para los ejes
eje_x = ["Gustvo Petro", "Rodolfo Hernandez"]
eje_y = [11.2, 10.5] #millones de personas

# creamos la grafica de barras
plt.bar(eje_x, eje_y)

#leyenda en el eje_y
plt.ylabel("No. Vts")

#leyenda en el eje_x
plt.xlabel("Candidatos a presidencia")

#titulo de la grafica
plt.title("Resultados de las elecciones presidenciales 2022")

#mostrar la grafica
plt.show()
```
